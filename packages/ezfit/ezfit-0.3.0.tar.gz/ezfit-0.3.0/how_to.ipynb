```python
import ezfit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def line(x, m, b):
    return m * x + b


test_data = pd.DataFrame(
    {
        "x": np.linspace(0, 10, 100),
        "y": np.linspace(0, 10, 100) + np.random.normal(0, 0.9, 100),
        "y_err": np.sqrt(np.linspace(0, 10, 100)) + 0.1,
    }
)
model, ax = test_data.fit(
    line,
    "x",
    "y",
    "y_err",
    m={"value": 1, "min": 0.8, "max": 1.2},
    b={"value": 0, "fixed": True},
)
print(model)
plt.show()
```
